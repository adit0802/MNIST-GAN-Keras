## GAN:-
Generative Adversarial Networks (GANs) are a powerful class of neural networks that are used for unsupervised learning. GANs are basically made up of a system of two competing neural network models which compete with each other and are able to analyze, capture and copy the variations within a dataset. GANs can create images that look like photographs of human faces, even though the faces don't belong to any real person.

## Step 3: Defining parameters to be used in later process
## Defining input shape


```python
image_shape = (28,28,1)
print(image_shape)
latent_dimensions = 10
```

# Step 4:Defining a utility function to build the Generator Model


```python
# Defining a utility function to build the Generator Model

def build_generator():

  model = Sequential() # stack layers one after another

  # Building the input layer
  model.add(Dense(7 * 7 * 128, activation='relu', input_dim=latent_dimensions))
  # the first layer is a densely connected (fully connected) layer with 128*8*8 neurons.
  # This layer takes input noise vector from the latent space, which has a length of 10000 (latent_dimensions), and transforms it into a higher dimesional representation with 8192 neurons (128*8*8)
  # relu introduces non-linearity in the model.

  model.add(Reshape((7,7,128)))
  # after the dense layer, the output is reshaped to form a 3d tensor of shape (8,8,128), this is done to prepare for the subsequent convolutional layers.
  # Reshape is used to convert the output 1d vector to 3d tensor
  # first dimension has a size of 8, representing the height of the tensor
  # second dimension has a size of 8, representing the width of the tensor
  # third dimension has a size of 128, representing the number of channels (feature maps) in the tensor.
  # 128 is the number of output images we get through 128 filters
  # In summary, the generator takes an 8x8 feature map with 128 channels as an intermediate representation, applies 128 filters (convolutional kernels) to it, and produces 128 corresponding output feature maps. These output feature maps are then further processed to generate the final image with the desired size specified in image_shape (e.g., 32x32).

  model.add(UpSampling2D())
  # upsampling 2d doubles the size of the image in both height and width dimensions.
  # By doing this, the UpSampling2D layer increases the spatial resolution of the image, making it larger. This process helps in recovering fine details that may have been lost during downsampling or in previous layers of the generator. It's like zooming in on the original image, but instead of creating new details, it replicates the existing information to fill in the gaps.
  # In the generator model, the UpSampling2D layers are used to gradually increase the resolution of the intermediate feature maps, allowing the generator to create higher-quality and more realistic images as it progresses through the network. This is essential for generating visually appealing and detailed images.

  model.add(Conv2D(128, kernel_size=3, padding="same"))
  # 128 is the no. of filters. Each filter learns different patterns or features in the input data.
  # kernel size is 3, meaning the filter will slide over input image and scan a 3x3 region at a time.
  # output size of the convolutional layer will be the same as the input size.
  # It pads the input feature map with zeros around the edges so that the filter can properly scan the entire input without going out of bounds.

  model.add(BatchNormalization(momentum=0.78))
  # Batch normalization is a technique that helps stabilize and accelerate the training process in deep neural networks.
  # It normalizes the activations of the previous layer for each batch of data, making the optimization process smoother and reducing the impact of vanishing or exploding gradients.
  # The momentum=0.78 parameter controls how fast the batch normalization adapts to changes during training. A value close to 1 means it adapts slowly, and a value closer to 0 means it adapts faster.

  model.add(Activation("relu"))
  # this activation function is applied to the output of the batchnormalization layer
  # relu introduces non linearity in the model, enabling it learn more complex patterns and make the generated images more realistic
  # it replaces negative pixel values with zero, and leaves positive values unchanged

  model.add(UpSampling2D())

  model.add(Conv2D(1, kernel_size=3, padding="same"))

  model.add(Activation("tanh"))
  # The "tanh" function squashes the pixel values in the output to the range [-1, 1].
  # For positive values of x, the function outputs a value close to 1.
  # For negative values of x, the function outputs a value close to -1.
  # For x close to 0, the output approaches 0.

  # Generating the output image
  noise = Input(shape=(latent_dimensions, ))
  # print('noise shape',noise.shape, latent_dimensions)
  # noise is the input to the generator, which represents the random noise vector from the latent space.
  # the shape of the input noise vector here is (10000,)
  # (100, ) is a tuple with one element 100, the comma is used to specify that 100 is element in a tuple.

  image = model(noise)
  # here an image is generated by passing the noise to the generator model.
  # forward pass involves passing the random noise vector through the layers of the generator to create a generated image.

  return Model(noise, image)
  # In Keras, a Model is a high-level class that allows you to define a model by specifying its input(s) and output(s).
  # It encapsulates the architecture of the neural network, making it easy to train, evaluate, and use the model for inference.
  # here, noise is the input and image is the output

```

## Step 5: Defining a utility function to build the Discriminator model

```python
# Defining a utility function to build Discriminator model
from keras.layers import *
#from keras.layers import LeakyRelu


def build_discriminator():
  # Building convnet
  model = Sequential()

  model.add(Conv2D(32, kernel_size=3, strides=2, padding="same", input_shape= image_shape))
  model.add(LeakyReLU(alpha=0.2))
  model.add(Dropout(0.25))

  model.add(Conv2D(64, kernel_size=3, strides=2, padding="same"))
  # model.add(ZeroPadding2D(padding=((0,1), (0,1))))
  model.add(BatchNormalization(momentum=0.82))
  model.add(LeakyReLU(alpha=0.25))
  model.add(Dropout(0.25))

  model.add(Conv2D(128, kernel_size=3, strides=2, padding="same"))
  model.add(BatchNormalization(momentum=0.82))
  model.add(LeakyReLU(alpha=0.25))
  model.add(Dropout(0.25))

  # model.add(Conv2D(256, kernel_size=3, strides=2, padding="same"))
  # model.add(BatchNormalization(momentum=0.82))
  # model.add(LeakyReLU(alpha=0.25))
  # model.add(Dropout(0.25))

  # Building output layer by using DNN
  model.add(Flatten())
  #model.add(Dense(128, activation="relu"))
  model.add(Dense(1, activation="sigmoid"))

  image = Input(shape=image_shape)
  validity = model(image)

  return Model(image, validity)
```

## Step 6:Defining a utility function to display the generated images

```python

def display_images():
  r, c = 4,4
  noise = np.random.normal(0,1, (r * c, latent_dimensions))
  generated_images = generator.predict(noise)


  # scaling the generated images
  generated_images = 0.5*generated_images + 0.5

  fig, axs = plt.subplots(r, c)
  count = 0
  for i in range(r):
    for j in range(c):
      axs[i,j].imshow(generated_images[count, :,:,])
      axs[i, j].axis('off')
      count +=1
  plt.show()
  plt.close()
```

## Step 7: Building the Generative Adversarial Network and compiling the discriminator

```python
discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer=Adam(0.0002,0.5), metrics=['accuracy'])

discriminator.trainable = False
```

## Step 8: Building the generator and checking the validity of the generated image

```python
# Building the generator

generator = build_generator()
z = Input(shape=(latent_dimensions, ))
image = generator(z)

# checking the validity of the generated image
valid = discriminator(image)
```

## Step 9: Defining the combined model of the Generator and Discriminator

```python
# Defining the combined model of the Generator and Discriminator

combined_network = Model(z, valid)

combined_network.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5))
```

##  Step 10: Final Step - Training the network

```python
# Final Step - Training the network

# num_epochs = 50
num_epochs = 500
batch_size =32
display_interval = 10
# display_interval = 20
losses=[]

# Normalizing the input data
X = (X/127.5) - 1.

# Defining the Adversarial ground truths
valid = np.ones((batch_size, 1))

# adding some noise

valid +=0.05 * np.random.random(valid.shape)
fake = np.zeros((batch_size, 1))
fake +=0.05*np.random.random(fake.shape)

for epoch in range(num_epochs):
  # training the discriminator
  # sampling a random half of images
  index = np.random.randint(0, X.shape[0], batch_size)
  images = X[index]

  # sampling noise and generating a batch of new images
  noise = np.random.normal(0,1, (batch_size, latent_dimensions))
  generated_images = generator.predict(noise)

  # Training the discriminator to detect more accurately whether a generated image is real or fake
  discm_loss_real = discriminator.train_on_batch(images, valid)
  discm_loss_fake = discriminator.train_on_batch(generated_images, fake)
  discm_loss = 0.5 * np.add(discm_loss_real, discm_loss_fake)

  # traing the generatorn to generat image
  genr_loss = combined_network.train_on_batch(noise, valid)

  # Tracking the progress
  if epoch % display_interval ==0:
    display_images()
```
